## **Matplotlib Scatter**

```python
xpoint=np.array([9,5,4,6,8,7,6,5,4,3])
ypoint=np.array([3,5,7,9,5,6,4,2,8,0])
plt.scatter(xpoint,ypoint)                  # Used to show only scattered points
plt.show()
```

```python
x=np.array([9,5,4,6,8,7,6,5,4,3])             # Plot 2 different scatter graphs and comparing them
y=np.array([3,5,7,9,5,6,4,2,8,0])
plt.scatter(x,y)
z=np.array([8,9,5,6,4,5,6,8,7,2])
w=np.array([5,6,7,3,4,8,6,1,0,5])
plt.scatter(z,w)                      
```

```python
x=np.array([9,5,4,6])                          # Display color of the points as specified
y=np.array([3,5,7,9])
c= np.array(['red','blue','green','purple'])
plt.scatter(x,y, c = c)
```

```python
x=np.array([1,4,7,10])                  # Marker's size is set
y=np.array([2,5,8,11])
size= np.array([100,200,300,400])
plt.scatter(x,y,s=size)
plt.show()
```

```python
x=np.array([10,4,7,10])
y=np.array([10,4,8,10])
size= np.array([100,200,300,400])
c= np.array(['black','blue','green','purple'])
plt.scatter(x,y,c=c,s=size,alpha=0.5)                  # alpha is used for visibility
plt.show()
```

```python
x=np.array([10,4,7,10])
y=np.array([10,4,8,10])
size= np.array([100,200,300,400])                                             # cmap= virdis,nipy_spectrum
c= np.array(['black','blue','green','purple'])
plt.scatter(x,y,c=c,s=size,alpha=0.5,cmap='nipy_spectrum')                    # cmap is used to create a colorbar
plt.colorbar()                                                                # alpha is used for visibility
plt.show()
```

## PIECHART

```python
y=np.array([1,2,3,4])
plt.pie(y)             # Used to make piechart
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
plt.pie(y,labels=mylabel)                            # Defines the pie chart with labelling
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
plt.pie(y,labels=mylabel, startangle=90)           # Provides a starting angle for 1st value
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
myexplode=[0.2,0,0,0]                                          # Used to divide the piechart in parts
plt.pie(y,labels=mylabel, startangle=90,explode = myexplode)
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
myexplode=[0.2,0,0,0]                                                             # Used to divide the piechart in parts
plt.pie(y,labels=mylabel, startangle=90,explode = myexplode,shadow=True)          # Shadow is used to convert it into 3D
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
myexplode=[0.2,0,0,0] 
mycolor= ['red','yellow','blue','orange']                                                                   # Used to divide the piechart in parts
plt.pie(y,labels=mylabel, startangle=90,explode = myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')  # colors used to specify color and autopct is used to show the parts in percentage
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
myexplode=[0.2,0,0,0] 
mycolor= ['red','yellow','blue','orange']                                                                 # Used to divide the piechart in parts
plt.pie(y,labels=mylabel, startangle=90,explode = myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%') # colors is used to specify color and autopct is used to show the parts in percentage
plt.legend()                                                                                              # Used to display the color block in piechart
```

```python
y=np.array([1,2,3,4])
mylabel=["apple","mango","banana","orange"]
myexplode=[0.2,0,0,0] 
mycolor= ['red','orange','yellow','green']                                                                  # Used to divide the piechart in parts
plt.pie(y,labels=mylabel, startangle=90,explode = myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%') # colors is used to specify color and autopct is used to show the parts in percentage
plt.legend(title="Color")                                                                                 # Used to display the color block with heading
```
