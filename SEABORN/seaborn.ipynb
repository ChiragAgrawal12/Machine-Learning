## RELATIONAL PLOT( relplot )

```python
sns.relplot(x= 'total_bill',y= 'tip', data=data)     # Display relplot of tip and total bill
```

```python
data['time'].unique()             # Displaying unique values for time
```

```python
sns.relplot(x= 'total_bill',y= 'tip', hue= 'time',data=data)    # Display customers for lunch and dinner with seperate colors and hue is used to show the relation according to the provided variable
```

```python
sns.relplot(x= 'total_bill', y= 'tip', hue= 'time', style= 'smoker',data=data)    # With the help of style, people are classified as smoker and no smoker
```

```python
sns.relplot(x= 'total_bill', y= 'tip', hue= 'time', style= 'smoker',data=data,kind= 'scatter',col='smoker')  # kind is used for type of plot and col is used to make columns
```

```python
sns.relplot(x= 'total_bill', y= 'tip', hue= 'time', style= 'smoker',data=data,kind='scatter',row='time')          # row is used to plot rows
```

```python
hii=(data['day']).unique()
hii
```

```python
sns.relplot(x= 'total_bill', y= 'tip', hue= 'time',data=data,kind= 'scatter',col='day')         # Graphs are printed according to day
```

```python
sns.relplot(x= 'total_bill', y= 'tip',data=data,)
plt.title("total bill vs tip",fontsize=20,color='blue')           # title of the graph is written
plt.xlabel("total bill",fontsize=20,color='blue')                 # label is provided to x
plt.ylabel("tip",fontsize=20,color='blue')                        # label is provided to y
```

# CATPLOT

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'swarm')    # Categorial graph is printed and kind is used for swarm style
```

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'strip')            # Strip plots individual points as scatter along a categorical axis.
```

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'box')          # box display the distribution of numerical data and identify outliers
```

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'bar')   # bar is used used to show comparisons between groups for maximum bill ranges
```

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'violin')    # It shows the distribution of a dataset as violin style

```

```python
sns.catplot(x='day', y='total_bill',data=data,kind= 'boxen')  # Improved version of the box plot designed for large datasets,useful when dealing with skewed or highly variable data
```
