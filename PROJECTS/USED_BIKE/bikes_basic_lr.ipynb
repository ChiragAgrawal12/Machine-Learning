### **LOSS FUNCTION**

#### **MSE**

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE=mean_squared_error(y_test,pred)
print("MSE :-",MSE)
```

#### **MAE**

```python
from sklearn.metrics import mean_absolute_error
```

```python
MAE= mean_absolute_error(y_test,pred)
print("MAE :-",MAE)
```

#### **RMSE**

```python
import math
import numpy as np
```

```python
print("RMSE :-",math.sqrt(mean_squared_error(y_test,pred)))
```

```python
print("By using numpy:-",np.sqrt(mean_squared_error(y_test,pred)))
```

#### **HUBER LOSS**

```python
print("Huber loss :-",mean_squared_error(y_test,pred)+ mean_absolute_error(y_test,pred))     # Or use print("Huber loss :-",MSE+MAE)
```

### MODEL EVALUATION

* Accuracy
* Precison
* Recall
* F1 Score

```python
pred=(pred>0.5).astype(int).ravel()
pred
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

#### **Confusion Matrix**

```python
print("Confusion Matrix :- \n",confusion_matrix(y_test,pred))     # Confusion matrix
```

```python
y_test1=(y_test>0.5).astype(int)              # Regression to Classification
y_test1
```

#### **Precision Score**

```python
print("Precision Score :-",precision_score(y_test1,pred))         # Precision score
```

#### **Recall Score**

```python
print("Recall Score :-",recall_score(y_test1,pred))              # Recall score
```

#### **F1 Score**

```python
print("F1 Score :-",f1_score(y_test1,pred))                     # F1 score
```
