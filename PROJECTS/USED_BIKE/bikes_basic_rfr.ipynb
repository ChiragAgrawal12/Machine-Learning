```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42) # Testsize how much data u want to keep for testing
```

#### **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
obj=RandomForestRegressor()
obj.fit(x_train,y_train)         # Fit the model
```

```python
print("Training Score: ",obj.score(x_train,y_train))     # Training score
print("Testing Score: ",obj.score(x_test,y_test))        # Testing score
```

```python
pred=obj.predict(x_test)        # Predict the value
pred
```

```python
obj_with_parameters=RandomForestRegressor(n_estimators=150,max_depth=2,min_samples_leaf=1,oob_score=True)
obj_with_parameters.fit(x_train,y_train)     # Fit the model
```

```python
print("Training Score: ",obj_with_parameters.score(x_train,y_train))     # Training score
print("Testing Score: ",obj_with_parameters.score(x_test,y_test))        # Testing score
```

```python
pred=obj_with_parameters.predict(x_test)      # Predict the value
pred
```
