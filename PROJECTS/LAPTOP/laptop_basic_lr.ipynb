### Now split the data

```python
x=df.drop(['Price'],axis=1)
y=df[['Price']]
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#### **LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()    # Object
print(type(lr))
```

```python
lr.fit(x_train,y_train)     # Train a model
```

```python
print("Training Score:",lr.score(x_train,y_train))
print("Testing Score:",lr.score(x_test,y_test))
```

```python
pred=lr.predict(x_test)     # Prediction about testing
pred
```

```python
x_test.shape
```

```python
pred,y_test
```

#### **LOSS FUNCTION** 

**MSE**

```python
from sklearn.metrics import mean_squared_error
print("MSE:",mean_squared_error(y_test,pred))
```

**MAE**

```python
from sklearn.metrics import mean_absolute_error
print("MAE:",mean_absolute_error(y_test,pred))
```

**RMSE**

```python
import math
print("RMSE:",math.sqrt(mean_squared_error(y_test,pred)))
```

```python
import numpy as np
print("RMSE:",np.sqrt(mean_squared_error(y_test,pred)))
```

**HUBER LOSS**

```python
print("HUBER LOSS:",mean_squared_error(y_test,pred)+mean_absolute_error(y_test,pred))
```

```python
pred=(pred>0.5).astype(int).ravel()       # Model evaluation
pred
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

**CONFUSION MATRIX**

```python
print("CONFUSION MATRIX\n",confusion_matrix(y_test,pred))
```
