encoding of cat column

```python
dt={
    'Loyal Customer':1,
    'disloyal Customer':0         # Dictonary
    }
#dict_customer_type.keys()
#dict_customer_type.values()
df['Customer Type']=df['Customer Type'].map(dt) # Used to map dict and feature
df['Customer Type'].unique()
```

```python
dt1={
    'Male':1,
    'Female':0                    # Dictonary
}
df['Gender']=df['Gender'].map(dt1)  # Used to map dict and feature
df['Gender'].unique()
```

```python
dt2={
    'Personal Travel':1,
    'Business travel':0             # Dictonary
}
df['Type of Travel']=df['Type of Travel'].map(dt2)  # Used to map dict and feature
df['Type of Travel'].unique()
```

```python
dt3={
    'Eco Plus':1,
    'Business':0,                # Dictonary
    'Eco':2
}
df['Class']=df['Class'].map(dt3)   # Used to map dict and feature
df['Class'].unique()
```

```python
dt4={
    'neutral or dissatisfied':1,
    'satisfied':0                             # Dictonary
}
df['satisfaction']=df['satisfaction'].map(dt4)  # Used to map dict and feature
df['satisfaction'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Customer Type'],axis=1)
x
```

```python
y=df['Customer Type']
y
```

```python
df.duplicated().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#### **DECISION TREE**

DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
obj=DecisionTreeClassifier() # Decision tree classifier object
obj.fit(x_train,y_train) # Fit the model with train data
```

```python
print("Training Score: ",obj.score(x_train,y_train)) # Training score
print("Testing Score: ",obj.score(x_test,y_test))    # Testing score
```

DECISION TREE REGRESSOR

```python
obj1=DecisionTreeRegressor() # Decision tree regressor object
obj1.fit(x_train,y_train) # Fit the model with train data
```

```python
print("Training Score: ",obj1.score(x_train,y_train)) # Training score
print("Testing Score: ",obj1.score(x_test,y_test))    # Testing score
```
